```python
total = []
pas = []
rush = []
offense = []
name_list = []

for item in ['THURSDAY', 'SUNDAY', 'WEDNESDAY']:
    
    src_path = 'Week-12/WEEK-12-DSA-STATS-' + item + '-CLEANED.xlsx'
    xls = xlrd.open_workbook(src_path, on_demand=True)
    sheet_names = xls.sheet_names()

    for sheet in sheet_names:
        df = pd.read_excel(src_path, sheet_name=sheet)
        for index in range(len(df)):
            if df['home'][index] != 0:
                home = df['Unnamed: 11'][index]
                break
        home_score = df['home'][len(df)-1]
        away_score = df['away'][len(df)-1]

        for qb in df[df['Unnamed: 12'] == 'QB']['stats...'].unique():
            QB = df[df['stats...'] == qb]
            name_list.append(qb)
            pass_score = round(QB[QB['Unnamed: 13'] == 'pass']['dsa'].sum()*0.6, 1)
            pas.append(pass_score)
            rush_score = round(QB[QB['Unnamed: 13'] == 'rush']['dsa'].sum(), 1)
            rush.append(rush_score)
            total_score = round(pass_score + rush_score, 1)
            total.append(total_score)
            if QB['Unnamed: 11'].iloc[0] == home:
                offense.append(str(int(total_score*100/home_score)) + ' %')
            else:
                offense.append(str(int(total_score*100/away_score)) + ' %')
```

```python
result = pd.DataFrame({'Name':name_list, 'Total':total, 'Pass':pas, 'Rush':rush, 'Offense':offense})
result = result.sort_values('Total', ascending=False)
result_10 = result[:10]
last = result.iloc[-1,:]
result_10 = result_10.append(last)
result_10
```
