```python
import pandas as pd
import numpy as np

pd.set_option('display.max_columns', 30)
```

```python
data = pd.read_excel('Week-13/WEEK-13-DSA-STATS-SUNDAY-CLEANED.xlsx', sheet_name='TEN-CLE')
```

```python
start_index = 148
end_index = 167

zeros = np.zeros((end_index-start_index, 28))
columns = ['#', 'Team1LogoPath', 'Team1BkgPath', 'Team1FirstName', 'Team1LastName', 'Team1Score',
          'Team2LogoPath', 'Team2BkgPath', 'Team2FirstName', 'Team2LastName', 'Team2Score', 
          'BallPossession', 'BallOnTeam', 'BallOnYards', 'YdsToTouchdown', 'Down', 'YardsToGo',
          'Quarter', 'Time', 'WindDirection', 'WindSpeed', 'SellScore', 'BuyScore', 'MeanScore',
          'Commentary', 'YellowLine', 'BallPosition', 'Temp']

gif = pd.DataFrame(zeros, columns=columns)
```

```python
teamlist = {'Arizona':['Cardinals', 'arizonaCardinals.png', 'Rectangle 741.png', 'ARI'],
            'Atlanta':['Falcons', 'atlantaFalcons.png', 'Rectangle 742.png', 'ATL'],
            'Baltimore':['Ravens', 'baltimoreRavens.png', 'Rectangle 752.png', 'BAL'],
            'Buffalo':['Bills', 'buffaloBills.png', 'Rectangle 747.png', 'BUF'],
            'Carolina':['Panthers', 'carolinaPanthers.png', 'Rectangle 755.png', 'CAR'],
            'Chicago':['Bears', 'chicagoBears.png', 'Rectangle 741.png', 'CHI'],
            'Cincinnati':['Bengals', 'cincinnatiBengals.png', 'Rectangle 758.png', 'CIN'],
            'Cleveland':['Browns', 'clevelandBrowns.png', 'Rectangle 758.png', 'CLE'],
            'Dallas':['Cowboys', 'dallasCowboys.png', 'Rectangle 755.png', 'DAL'],
            'Denver':['Broncos', 'denverBroncos.png', 'Rectangle 758.png', 'DEN'],
            'Detroit':['Lions', 'detroitLions.png', 'Rectangle 746.png', 'DET'],
            'Green Bay':['Packers', 'greenBayPackers.png', 'Rectangle 765.png', 'GB'],
            'Houston':['Texans', 'houstonTexans.png', 'Rectangle 765.png', 'HOU'],
            'Indianapolis':['Colts', 'indianapolisColts.png', 'Rectangle 765.png', 'IND'],
            'Jacksonville':['Jaguars', 'jacksonvilleJaguars.png', 'Rectangle 759.png', 'JAC'],
            'Kansas City':['Chiefs', 'kansasCityChiefs.png', 'Rectangle 744.png', 'KC'],
            'Las Vegas':['Raiders', 'lasVegasRaiders.png', 'Rectangle 768.png', 'LV'],
            'Los Angeles Chargers':['Chargers', 'losAngelesChargers.png', 'Rectangle 755.png', 'LAC'],
            'Los Angeles Rams':['Rams', 'losAngelesRams.png', 'Rectangle 755.png', 'LA'],
            'Miami':['Dolphins', 'miamiDolphins.png', 'Rectangle 751.png', 'MIA'],
            'Minnesota':['Vikings', 'minnesotaVikings.png', 'Rectangle 770.png', 'MIN'],
            'New England':['Patriots', 'newEnglandPatriots.png', 'Rectangle 750.png', 'NE'],
            'New Orleans':['Saints', 'newOrleansSaints.png', 'Rectangle 771.png', 'NO'],
            'New York Giants':['Giants', 'newYorkGiants.png', 'Rectangle 746.png', 'NYG'],
            'New York Jets':['Jets', 'newYorkJets.png', 'Rectangle 769.png', 'NYJ'],
            'Philadelphia':['Eagles', 'philadelphiaEagles.png', 'Rectangle 768.png', 'PHI'],
            'Pittsburgh':['Steelers', 'pittsburghSteelers.png', 'Rectangle 754.png', 'PIT'],
            'San Francisco':['49ers', '49ers.png', 'Rectangle 741.png', 'SF'],
            'Seattle':['Seahawks', 'seattleSeahawks.png', 'Rectangle 757.png', 'SEA'],
            'Tampa Bay':['Buccaneers', 'tampaBayBuccaneers.png', 'Rectangle 745.png', 'TB'],
            'Tennessee':['Titans', 'tennesseeTitans.png', 'Rectangle 770.png', 'TEN'],
            'Washington':['Washington Team', 'washingtonRedskins.png', 'Rectangle 760.png', 'WAS']   
           }
```

```python
df = data.iloc[start_index:end_index, :]
df = df.reset_index(drop=True)

team1 = 'Tennessee'
team2 = 'Cleveland'
balpossession = 'Team1'
temp = '11°'
winddirection = 'Against'
windspeed = '11 mph'

downdic = {0: 'th', 1: 'st', 2:'nd', 3:'rd', 4:'th'}
quarterdic = {1: 'st', 2:'nd', 3:'rd', 4:'th', 5:'th'}
quarter = quarterdic[df['quarter'][0]]

gif['Team1LogoPath'] = teamlist[team1][1]
gif['Team2LogoPath'] = teamlist[team2][1]
gif['Team1BkgPath'] = teamlist[team1][2]
gif['Team2BkgPath'] = teamlist[team2][2]
gif['Team1FirstName'] = team1.upper()
gif['Team2FirstName'] = team2.upper()
gif['Team1LastName'] = teamlist[team1][0].upper()
gif['Team2LastName'] = teamlist[team2][0].upper()
gif['BallPossession'] = balpossession
gif['Temp'] = temp
gif['WindDirection'] = winddirection
gif['WindSpeed'] = windspeed

for index in range(len(df)):

    gif['#'][index] = int(index+1)
    gif['MeanScore'][index] = df['curr_play_ds'][index]
    gif['BuyScore'][index] = df['curr_play_ds'][index] * 1.03
    gif['SellScore'][index] = df['curr_play_ds'][index] * 0.97
    gif['Commentary'][index] = df['description'][index]
    gif['Down'][index] = str(df['down'][index]) + downdic[df['down'][index]]
    gif['YdsToTouchdown'][index] = int(df['ytg_td'][index])
    gif['YellowLine'][index] = int(df['ytg_td'][index])
    gif['BallPosition'][index] = int(df['ytg_td'][index])
    gif['YardsToGo'][index] = int(df['ytg'][index])
    gif['Quarter'][index] = str(df['quarter'][index]) + quarter
    gif['Team1Score'][index] = int(df['home'][index])
    gif['Team2Score'][index] = int(df['away'][index])
    
    if 'TOUCHDOWN' in gif['Commentary'][index]:
        gif['Down'][index] = 'Touchdown'
    
    if 'field goal is GOOD' in gif['Commentary'][index]:
        gif['Down'][index] = 'Field Goal'
    
    if 'extra point is GOOD' in gif['Commentary'][index]:
        gif['Down'][index] = 'Attempt Good'
        gif['YdsToTouchdown'][index] = gif['YdsToTouchdown'][index-1]
        gif['YellowLine'][index] = gif['YellowLine'][index-1]
        gif['BallPosition'][index] = gif['BallPosition'][index-1]
        
    if 'TWO-POINT CONVERSION' in gif['Commentary'][index]:
        gif['Down'][index] = 'Attempt Good'
        gif['YdsToTouchdown'][index] = gif['YdsToTouchdown'][index-1]
        gif['YellowLine'][index] = gif['YellowLine'][index-1]
        gif['BallPosition'][index] = gif['BallPosition'][index-1]

# BallOnTeam and BallOnYards
if gif['BallPossession'][0] == 'Team1':
    ballonteam1 = team1
    ballonteam2 = team2
else:
    ballonteam1 = team2
    ballonteam2 = team1

for index in range(len(df)):
    if gif['YdsToTouchdown'][index] > 50:
        gif['BallOnYards'][index] = 100 - gif['YdsToTouchdown'][index]
        gif['BallOnTeam'][index] = teamlist[ballonteam1][3]
    else:
        gif['BallOnYards'][index] =  gif['YdsToTouchdown'][index]
        gif['BallOnTeam'][index] = teamlist[ballonteam2][3]
    
# Time
for index in range(len(df)):
    if gif['Commentary'][index][0] == '(':
        gif['Time'][index] = str(gif['Commentary'][index].split(' ')[0][1:-1])

if gif['Time'][0] == 0:
    gif['Time'][0] = gif['Time'][1]

if gif['Time'][gif.index[-1]] == 0:
    gif['Time'][gif.index[-1]] = gif['Time'][gif.index[-2]]
```

```python
gif.to_excel('Week-13/Titans.xlsx', index=False)
```
