```python
import pandas as pd 
import numpy as np
import plotly
import plotly.express as px
import plotly.graph_objects as go
```

```python
df_cases = pd.read_csv('https://raw.githubusercontent.com/CSSEGISandData/COVID-19/master/csse_covid_19_data/csse_covid_19_time_series/time_series_covid19_confirmed_global.csv')
df_deaths = pd.read_csv('https://raw.githubusercontent.com/CSSEGISandData/COVID-19/master/csse_covid_19_data/csse_covid_19_time_series/time_series_covid19_deaths_global.csv')
```

```python
df_cases.sort_values(df_cases.columns[-1], ascending = False).head(15)
```

```python
# Preprocessing
# Cases dataset

df_cases = df_cases[df_cases['Province/State'].isnull()]
df_cases = df_cases.drop('Province/State', axis = 1)
df_cases = df_cases.drop(df_cases[df_cases['Country/Region'] == 'Diamond Princess'].index, axis = 0)
df_cases = df_cases.drop(df_cases[df_cases['Country/Region'] == 'MS Zaandam'].index, axis = 0)
df_cases['Country/Region'] = df_cases['Country/Region'].replace('Taiwan*', 'Taiwan')

# Deaths dataset

df_deaths = df_deaths[df_deaths['Province/State'].isnull()]
df_deaths = df_deaths.drop('Province/State', axis = 1)
df_deaths = df_deaths.drop(df_deaths[df_deaths['Country/Region'] == 'Diamond Princess'].index, axis = 0)
df_cases = df_cases.drop(df_cases[df_cases['Country/Region'] == 'MS Zaandam'].index, axis = 0)
df_deaths['Country/Region'] = df_deaths['Country/Region'].replace('Taiwan*', 'Taiwan')
```

```python
df_cases
```

```python
# Daily new infections

daily_sum = []
summ = 0
for i in df_cases.columns[-200:]:
    daily_sum.append(df_cases[i].sum() - summ)
    summ = df_cases[i].sum()
    
fig = px.bar(x = df_cases.columns[-199:], y = daily_sum[1:], color = daily_sum[1:], height = 550)
fig.update_layout(title_text = 'Daily new cases around the world')
fig.update_layout(coloraxis_colorbar = dict(title = 'Daily new infections'))
fig.update_xaxes(title = 'Date', nticks = 20, tickangle = 35)
fig.update_yaxes(title = 'Daily new infections')
fig.show()
fig.write_image("daily_cases.png", width=1100, height=600)
#plotly.offline.plot(fig, filename='daily_cases.html')
#fig.write_html("file.html")
```

```python
# Daily new deaths

daily_sum = []
summ = 0
for i in df_deaths.columns[-200:]:
    daily_sum.append(df_deaths[i].sum() - summ)
    summ = df_deaths[i].sum()

fig = px.bar(x = df_deaths.columns[-199:], y = daily_sum[1:], color = daily_sum[1:], 
             height = 550, color_continuous_scale = 'Viridis')
fig.update_layout(title_text = 'Daily new deaths around the world')
fig.update_layout(coloraxis_colorbar = dict(title = 'Daily new deaths'))
fig.update_xaxes(title = 'Date', nticks = 20, tickangle = 35)
fig.update_yaxes(title = 'Daily new deaths')
fig.show()
fig.write_image("daily_deaths.png", width=1100, height=600)
```

```python
# Bar Chart

df_cases_sorted = df_cases.sort_values(df_cases.columns[-1], ascending = False)
countries = df_cases_sorted['Country/Region'].tolist()

fig = go.Figure(data = [go.Bar(x = countries[:30], y = df_cases_sorted.iloc[:30, -1].values)])
fig.update_layout(title_text = 'Coronavirus infections as of ' + df_cases.columns[-1])
fig.update_xaxes(tickangle = 45)
fig.show()
```

```python
# Bar Chart

df_deaths_sorted = df_deaths.sort_values(df_deaths.columns[-1], ascending = False)
countries = df_deaths_sorted['Country/Region'].tolist()

fig = go.Figure(data = [go.Bar(x = countries[:30], y = df_deaths_sorted.iloc[:30, -1].values)])
fig.update_layout(title_text = 'Coronavirus deaths as of ' + df_cases.columns[-1])
fig.update_xaxes(tickangle = 45)
fig.show()
```

```python
# Stacked Bar Chart

top_countries = ['US', 'Brazil', 'Russia', 'India', 'United Kingdom', 'Peru']
top_countries_list = ['US', 'Brazil', 'Russia', 'India', 'United Kingdom', 'Peru']
for j in range(6):
    top_countries_list[j] = []
    for i in range(-2, -31, -2):
        top_countries_list[j].append(df_cases[df_cases['Country/Region'] == top_countries[j]].iloc[0, i])
    top_countries_list[j] = list(reversed(top_countries_list[j]))

dates = df_cases.columns[-30::2]

fig = go.Figure(data = [go.Bar(name = top_countries[i], x = dates, y = top_countries_list[i]) 
                      for i in range(6)])
fig.update_layout(barmode = 'stack')
fig.update_layout(title_text = 'Coronavirus infections in top 6 countries')
fig.update_xaxes(title = 'Date')
fig.show()
fig.write_image("top_countries.png", width=1100, height=600)
```

```python
# Stacked Bar Chart

top_countries = ['US', 'Brazil', 'United Kingdom', 'Italy', 'France', 'Spain']
top_countries_list = ['US', 'Brazil', 'United Kingdom', 'Italy', 'France', 'Spain']
for j in range(6):
    top_countries_list[j] = []
    for i in range(-2, -31, -2):
        top_countries_list[j].append(df_deaths[df_deaths['Country/Region'] == top_countries[j]].iloc[0, i])
    top_countries_list[j] = list(reversed(top_countries_list[j]))

dates = df_deaths.columns[-30::2]

fig = go.Figure(data = [go.Bar(name = top_countries[i], x = dates, y = top_countries_list[i]) 
                      for i in range(6)])
fig.update_layout(barmode = 'stack')
fig.update_layout(title_text = 'Coronavirus deaths in top 6 countries')
fig.update_xaxes(title = 'Date')
fig.show()
```

```python
# Covid-19 cases in Hungary

hungary = df_cases[df_cases['Country/Region'] == 'Hungary']
daily_sum = []
summ = 0
for i in hungary.columns[-200:]:
    daily_sum.append(hungary[i].iloc[0] - summ)
    summ = hungary[i].iloc[0]

fig = px.bar(x = hungary.columns[-199:], y = daily_sum[1:], color = daily_sum[1:], height = 550)
fig.update_layout(title_text = 'Daily new cases in Hungary')
fig.update_layout(coloraxis_colorbar = dict(title = 'Daily new infections'))
fig.update_xaxes(title = 'Date', nticks = 20, tickangle = 35)
fig.update_yaxes(title = 'Daily new infections')
fig.show()
fig.write_image("hungary_daily_cases.png", width=1100, height=600)
```
