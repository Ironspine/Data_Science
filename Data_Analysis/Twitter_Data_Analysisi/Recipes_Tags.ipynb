```python
import pandas as pd
import string
import re
```

```python
df = pd.read_excel("Recipes Tags.xlsx")
```

```python
df
```

```python
# get only the score of ratings, all rating has a scale of 4 so that can be deleted
df['Rating'] = df['Rating'].apply(lambda x: x[:-2])
```

```python
print(df['ActiveTime'].isnull().sum())
print(df['TotalTime'].isnull().sum())

# I dropped these columns because they are all NaN valus, which does not make much sense
df = df.drop(['ActiveTime', 'TotalTime'], axis=1)
```

```python
def clean_data(x):
    
    x = x.translate(str.maketrans('', '', string.punctuation)) # remove punctuations
    x = x.translate(str.maketrans({'\r': '', '#': '', '\t': '', '\n': ''})) # replace tabs, new lines etc.
    pattern = re.compile(r'\s+')
    x = re.sub(pattern, ' ', x) # remove whitespaces
    return x.lower()
```

```python
df['Name'] = df['Name'].apply(clean_data)
df['Description'] = df['Description'].apply(clean_data)
df['Ingredients'] = df['Ingredients'].apply(clean_data)
df['Method'] = df['Method'].apply(clean_data)

df['Tags'] = df.fillna({'Tags':''}) # there are NaN values so they are replaced with an empty '' string
df['Tags'] = df['Tags'].apply(clean_data)
```
