```python
import pandas as pd
import numpy as np

pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', 200)
```

```python
data = pd.read_excel('Data_Oct_2020.xlsx')
```

```python
Sub_189 = data[(data['Visa Type'] == '189PTS Points-Tested Stream') & (data['EOI Status'] == 'SUBMITTED')]
Sub_189 = Sub_189.reset_index(drop=True)
Inv_189 = data[(data['Visa Type'] == '189PTS Points-Tested Stream') & (data['EOI Status'] == 'INVITED')]
Inv_189 = Inv_189.reset_index(drop=True)

Sub_190 = data[(data['Visa Type'] == '190SAS Skilled Australian Sponsored') & (data['EOI Status'] == 'SUBMITTED')]
Sub_190 = Sub_190.reset_index(drop=True)
Inv_190 = data[(data['Visa Type'] == '190SAS Skilled Australian Sponsored') & (data['EOI Status'] == 'INVITED')]
Inv_190 = Inv_190.reset_index(drop=True)

Sub_491_Family = data[(data['Visa Type'] == '491FSR Family Sponsored - Regional') & (data['EOI Status'] == 'SUBMITTED')]
Sub_491_Family = Sub_491_Family.reset_index(drop=True)
Inv_491_Family = data[(data['Visa Type'] == '491FSR Family Sponsored - Regional') & (data['EOI Status'] == 'INVITED')]
Inv_491_Family = Inv_491_Family.reset_index(drop=True)

Sub_491_State = data[(data['Visa Type'] == '491SNR State or Territory Nominated - Regional') & (data['EOI Status'] == 'SUBMITTED')]
Sub_491_State = Sub_491_State.reset_index(drop=True)
Inv_491_State = data[(data['Visa Type'] == '491SNR State or Territory Nominated - Regional') & (data['EOI Status'] == 'INVITED')]
Inv_491_State = Inv_491_State.reset_index(drop=True)
```

```python
df = pd.read_excel('SkillSelect_Sample.xlsx', index_col=0)
df['ANZSCO'] = df['ANZSCO'].astype('str')
df
```

```python
# POINTS

print('189 Submitted')
for num in range(65, 140, 5):
    EOI = Sub_189[Sub_189['Points'] == int(num)]
    for row in range(len(EOI)):
        if EOI['Occupation'].iloc[row].split()[0] in list(df['ANZSCO']):
            for index in df[df['ANZSCO'] == EOI['Occupation'].iloc[row].split()[0]].index:
                df['189 - EOI Sub ' + str(num) + ' Points'].iloc[index] = EOI['Count EOIs'].iloc[row]
    print(df['189 - EOI Sub ' + str(num) + ' Points'].isnull().sum())
print('')

print('189 Invited')
for num in range(65, 140, 5):
    EOI = Inv_189[Inv_189['Points'] == int(num)]
    for row in range(len(EOI)):
        if EOI['Occupation'].iloc[row].split()[0] in list(df['ANZSCO']):
            for index in df[df['ANZSCO'] == EOI['Occupation'].iloc[row].split()[0]].index:
                df['189 - Invited ' + str(num) + ' Points'].iloc[index] = EOI['Count EOIs'].iloc[row]
    print(df['189 - Invited ' + str(num) + ' Points'].isnull().sum())
print('')

print('190 Submitted')
for num in range(65, 130, 5):
    EOI = Sub_190[Sub_190['Points'] == int(num)]
    for row in range(len(EOI)):
        if EOI['Occupation'].iloc[row].split()[0] in list(df['ANZSCO']):
            for index in df[df['ANZSCO'] == EOI['Occupation'].iloc[row].split()[0]].index:
                df['190 - EOI Sub ' + str(num) + ' Points'].iloc[index] = EOI['Count EOIs'].iloc[row]
    print(df['190 - EOI Sub ' + str(num) + ' Points'].isnull().sum())
print('')

print('190 Invited')
for num in range(65, 130, 5):
    EOI = Inv_190[Inv_190['Points'] == int(num)]
    for row in range(len(EOI)):
        if EOI['Occupation'].iloc[row].split()[0] in list(df['ANZSCO']):
            for index in df[df['ANZSCO'] == EOI['Occupation'].iloc[row].split()[0]].index:
                df['190 - Invited ' + str(num) + ' Points'].iloc[index] = EOI['Count EOIs'].iloc[row]
    print(df['190 - Invited ' + str(num) + ' Points'].isnull().sum())
print('')

print('491 Submitted')
for num in range(65, 140, 5):
    EOI = Sub_491_State[Sub_491_State['Points'] == int(num)]
    for row in range(len(EOI)):
        if EOI['Occupation'].iloc[row].split()[0] in list(df['ANZSCO']):
            for index in df[df['ANZSCO'] == EOI['Occupation'].iloc[row].split()[0]].index:
                df['491 - EOI Sub ' + str(num) + ' Points'].iloc[index] = EOI['Count EOIs'].iloc[row]
    print(df['491 - EOI Sub ' + str(num) + ' Points'].isnull().sum())
print('')

print('491 Invited')
for num in range(65, 140, 5):
    EOI = Inv_491_State[Inv_491_State['Points'] == int(num)]
    for row in range(len(EOI)):
        if EOI['Occupation'].iloc[row].split()[0] in list(df['ANZSCO']):
            for index in df[df['ANZSCO'] == EOI['Occupation'].iloc[row].split()[0]].index:
                df['491 - Invited ' + str(num) + ' Points'].iloc[index] = EOI['Count EOIs'].iloc[row]
    print(df['491 - Invited ' + str(num) + ' Points'].isnull().sum())
print('')

print('491FS Submitted')
for num in range(65, 125, 5):
    EOI = Sub_491_Family[Sub_491_Family['Points'] == int(num)]
    for row in range(len(EOI)):
        if EOI['Occupation'].iloc[row].split()[0] in list(df['ANZSCO']):
            for index in df[df['ANZSCO'] == EOI['Occupation'].iloc[row].split()[0]].index:
                df['491FS - EOI Sub ' + str(num) + ' Points'].iloc[index] = EOI['Count EOIs'].iloc[row]
    print(df['491FS - EOI Sub ' + str(num) + ' Points'].isnull().sum())
print('')

print('491FS Invited')
for num in range(65, 125, 5):
    EOI = Inv_491_Family[Inv_491_Family['Points'] == int(num)]
    for row in range(len(EOI)):
        if EOI['Occupation'].iloc[row].split()[0] in list(df['ANZSCO']):
            for index in df[df['ANZSCO'] == EOI['Occupation'].iloc[row].split()[0]].index:
                df['491FS - Invited ' + str(num) + ' Points'].iloc[index] = EOI['Count EOIs'].iloc[row]
    print(df['491FS - Invited ' + str(num) + ' Points'].isnull().sum())
```

```python
for index in range(len(df)):
    if df['ANZSCO Occupations & Sub-specialisations'][index][:6].isnumeric():
        df['ANZSCO Occupations & Sub-specialisations'][index] = df['ANZSCO Occupations & Sub-specialisations'][index][6:]

df.insert(1, 'ANZSCO2', df['ANZSCO'])
df = df.drop(['ANZSCO'], axis=1)
df.rename(columns={'ANZSCO2': 'ANZSCO'}, inplace=True)

anzsco = []
for index in range(len(df)):
    if df['ANZSCO'][index] not in anzsco:
        anzsco.append(df['ANZSCO'][index])
    else:
        df['ANZSCO'][index] = ''
```
